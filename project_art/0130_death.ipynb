```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import numpy as np

# 폰트 경로
font_path = 'C:/Windows/Fonts/malgun.ttf'  # Malgun Gothic 폰트 경로
font_prop = fm.FontProperties(fname=font_path)

# 폰트 적용
plt.rcParams['font.family'] = font_prop.get_name()

# 데이터 불러오기
file_path = "박서보.xlsx"
df = pd.read_excel(file_path)

# 칼럼 이름의 공백 제거 (혹시 공백이 있을 경우)
df.columns = df.columns.str.strip()

# 가격을 정수로 변환 (숫자와 천 단위 구분자 , 제거)
df['price'] = df['price'].replace({',': ''}, regex=True).astype(float)

# 'date' 칼럼을 datetime 형식으로 변환
df['date'] = pd.to_datetime(df['date'])

# 연도와 월 추출 (예: 2024-12)
df['year_month'] = df['date'].dt.to_period('M').astype(str)

# 연도와 월별로 그룹화하고 평균 가격 계산
monthly_avg_price = df.groupby(['year_month', 'category_size'])['price'].mean().reset_index()

# 스타일 설정
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# 색상 설정 (category_size별로)
sns.scatterplot(data=monthly_avg_price, x='year_month', y='price', hue='category_size', palette='Set1', s=100)

# 제목과 레이블 설정
plt.title('Average Price by Month', fontsize=16)
plt.xlabel('Date (Year-Month)', fontsize=12)
plt.ylabel('Average Price', fontsize=12)

# x축 레이블 세로로 설정
plt.xticks(rotation=90)

# 범례 추가
plt.legend(title='Size Category', loc='upper left')

# 그래프 표시
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# 가격을 원화 형식으로 변환하는 함수
def to_won(x, pos):
    return f"₩{x:,.0f}"  # 천 단위 구분 기호와 원화 기호 추가


# 데이터 불러오기
file_path = "박서보.xlsx"
df = pd.read_excel(file_path)

# 칼럼 이름의 공백 제거
df.columns = df.columns.str.strip()

# 'Medium' 사이즈 데이터만 필터링
medium_df = df[df['category_size'] == 'medium']

# 날짜를 datetime 형식으로 변환
medium_df['date'] = pd.to_datetime(medium_df['date'])

# 가격을 정수로 변환 (천 단위 구분자 제거)
medium_df['price'] = medium_df['price'].replace({',': ''}, regex=True).astype(float)

# 월별 평균 가격 계산
monthly_avg_price = medium_df.groupby(medium_df['date'].dt.to_period('M'))['price'].mean().reset_index()

# 월별 거래량 계산
monthly_sales = medium_df.groupby(medium_df['date'].dt.to_period('M'))['price'].count().reset_index()
monthly_sales.rename(columns={'price': 'sales_count'}, inplace=True)

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(12, 6))

# 첫 번째 그래프: 가격 변화 (선 그래프)
monthly_avg_price['date'] = monthly_avg_price['date'].dt.strftime('%Y-%m-%d')
sns.lineplot(data=monthly_avg_price, x='date', y='price', marker='o', color='blue', ax=ax1)
ax1.yaxis.set_major_formatter(FuncFormatter(to_won))
ax1.set_xlabel('Month', fontsize=12)
ax1.set_ylabel('Average Price (₩)', fontsize=12)
ax1.set_title('Average Price & Sales Count for Medium Size', fontsize=16)

# 두 번째 그래프: 거래량 (바 그래프)
ax2 = ax1.twinx()  # x축을 공유하는 두 번째 y축 생성
sns.barplot(data=monthly_sales, x='date', y='sales_count', color='orange', ax=ax2, alpha=0.6)
ax2.set_ylabel('Sales Count', fontsize=12)

# 레이블 회전
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)

# 그래프 표시
plt.tight_layout()
plt.show()


#파란색 선 그래프는 주로 시간에 따른 가격의 변화를 나타냅니다.
#노란색 막대 그래프는 범주형 데이터(예: 크기별 거래량 또는 평균 가격)를 나타냅니다.
```

```python
# 지피티에게 코드 받아온 결과

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 🔹 1. 데이터 불러오기
file_path = "C:\src\data_python\박서보.xlsx"
df = pd.read_excel(file_path, sheet_name="Sheet1")

# 🔹 2. 컬럼 정리 (필요한 컬럼만 선택)
df_cleaned = df[['date', 'price', 'category_size']].copy()
df_cleaned.rename(columns={'date': 'transaction_date'}, inplace=True)

# 🔹 3. 데이터 타입 변환 및 NaN 처리
df_cleaned['transaction_date'] = pd.to_datetime(df_cleaned['transaction_date'], errors='coerce')
df_cleaned = df_cleaned.dropna(subset=['category_size'])

# 🔹 4. 2022년 10월 기준으로 데이터 나누기
cutoff_date = pd.Timestamp("2022-10-01")
df_before = df_cleaned[df_cleaned['transaction_date'] < cutoff_date]
df_after = df_cleaned[df_cleaned['transaction_date'] >= cutoff_date]

# 🔹 5. 거래량 및 평균 가격 계산
summary_before = df_before.groupby('category_size').agg(
    transaction_count=('price', 'count'),
    avg_price=('price', 'mean')
).reset_index()

summary_after = df_after.groupby('category_size').agg(
    transaction_count=('price', 'count'),
    avg_price=('price', 'mean')
).reset_index()

# 🔹 6. 거래량 및 가격 변화율 계산
summary_comparison = summary_before.merge(summary_after, on='category_size', suffixes=('_before', '_after'))
summary_comparison['transaction_count_change (%)'] = (
    (summary_comparison['transaction_count_after'] - summary_comparison['transaction_count_before']) 
    / summary_comparison['transaction_count_before']
) * 100

summary_comparison['avg_price_change (%)'] = (
    (summary_comparison['avg_price_after'] - summary_comparison['avg_price_before']) 
    / summary_comparison['avg_price_before']
) * 100

# 🔹 7. 시각화 데이터 준비
categories = summary_comparison['category_size']
count_before = summary_comparison['transaction_count_before']
count_after = summary_comparison['transaction_count_after']
price_before = summary_comparison['avg_price_before']
price_after = summary_comparison['avg_price_after']
price_change = summary_comparison['avg_price_change (%)']

# 🔹 8. 거래량 변화 (Bar Chart)
plt.figure(figsize=(10, 5))
bar_width = 0.35
x = np.arange(len(categories))

plt.bar(x - bar_width/2, count_before, bar_width, label="Before (Pre-Oct 2022)", alpha=0.7, color='blue')
plt.bar(x + bar_width/2, count_after, bar_width, label="After (Post-Oct 2022)", alpha=0.7, color='red')

plt.xticks(x, categories)
plt.xlabel("Size Category")
plt.ylabel("Transaction Count")
plt.title("Change in Artwork Transaction Count (Before vs After Oct 2022)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# 🔹 9. 평균 가격 변화 (Line + Bar Chart)
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.bar(categories, price_change, color='orange', alpha=0.6, label="Price Change (%)")
ax1.set_ylabel("Price Change (%)", color='orange')
ax1.set_xlabel("Size Category")
ax1.set_title("Average Artwork Price Change (Before vs After Oct 2022)")
ax1.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Baseline

# 🔹 Line plot for Before & After price comparison
ax2 = ax1.twinx()
ax2.plot(categories, price_before, marker='o', color='blue', label="Avg Price Before (₩)")
ax2.plot(categories, price_after, marker='o', color='red', label="Avg Price After (₩)")
ax2.set_ylabel("Average Price (₩)")

ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
